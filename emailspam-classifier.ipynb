```python
stopword = nltk.corpus.stopwords.words('english')
```

```python
lemmatizer=WordNetLemmatizer()
corpus=[]
for i in range(len(data)):
    review=re.sub('[^a-zA-Z0-9]',' ',data['MESSAGE'][i])  # removing all the character other than alphabets
    review=review.lower()                                 # Lowering all the cases
    review=review.split()                                 # spliting the sentence -> words for lemmatization
    review=[lemmatizer.lemmatize(word) for word in review if word not in set(stopwords.words('english'))]      # removing stopwords & get the word in it stem
    review=' '.join(review)                                # joining the words to rearrage to form the sent without stop words
    corpus.append(review)                                  # Appending to corpus
```

Modeling:

1- Logistic Regression:

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix
```

```python
# Converting the Words to Vector using Bag of words

from sklearn.feature_extraction.text import CountVectorizer

cv=CountVectorizer(max_features=2500,ngram_range=(1,3)) # top 2500 features are taken

X=cv.fit_transform(corpus).toarray()

y=data['CATEGORY']
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1,stratify=y)
```

```python
# create model1 (LogisticRegression) --->> Classification

model1=linear_model.LogisticRegression()
model1.fit(x_train,y_train)
```

```python
y_predict=model1.predict(x_test)
y_predict[:15]
```

```python
from sklearn import metrics
import sklearn

# Acc:

print(f"accuracy --->>> {metrics.accuracy_score(y_test,y_predict)*100:0.2f}%")
print(f"accuracy of errors --->>>{ 100-(metrics.accuracy_score(y_test,y_predict))*100:0.2f}%")

# confusion:

cm=metrics.confusion_matrix(y_test,y_predict)
tp,tn,fp,fn= cm[1,1],cm[0,0],cm[0,1],cm[1,0]
print("------------------------------------")
print("True positve prediction :",tp)
print("True Negative predicion :",tn)
print("False positve prediction:",fp)
print("False Negative prediction:",fn)
print("------------------------------------")
# Report:

class_report = classification_report(y_test, y_predict)
print("Classification Report:\n", class_report)
```
