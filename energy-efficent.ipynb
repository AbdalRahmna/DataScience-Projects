```python
class Var_Selector(BaseEstimator,TransformerMixin):

    def __init__(self,num_cols):
        self.num_cols=None
    def fit(self,data):
        self.num_cols=data.select_dtypes("number").columns
        return self
    def transform(self,data):
        selector=VarianceThreshold(threshold=0.01)
        selector.fit(data[self.num_cols])

        removed_cols=[column for column in self.num_cols if column not in \
                     self.num_cols[selector.get_support(indices=True)]]
        data=data.drop(columns=removed_cols)

        # remove high-low cardinality categorical:-

        removed_cols=data.nunique()[((data.nunique()==1)|data.nunique()==len(data))].index

        return data.drop(columns=removed_cols)
```

```python
# There No Any outliers :  :
features=data.select_dtypes(include="number").columns

for col in features:
    Q1_col,Q3_col=data[col].quantile([0.25,0.75])
    iqr=Q3_col-Q1_col
    low_limit=Q1_col-1.5*iqr
    upper_limit=Q3_col+1.5*iqr
    outlier=[]
    for x in data[col]:
         if ((x> upper_limit) or (x<low_limit)):
             outlier.append(x)
    if len(outlier)==0:
        print(f" * -- >> there No outlier in {col} feature")
    else:
        print(f"Ther outlier in this feature {col}")

    print(f"Q1 of {col} --->>> {Q1_col} \n Q3 of {col} ---->>> {Q3_col} \n iqr--->>{iqr}\n low_limit--->>>{low_limit} \n  upper_limit--->>> {upper_limit} \n outlier---->>> {outlier} \n Number of outliers --->>> {len(outlier)}")
    print("-"*25)
```

```python


train, test = train_test_split(data, test_size=0.2)
train_stats = train.describe()



train_stats.pop('Y1')
train_stats.pop('Y2')

train_stats = train_stats.transpose()
train_Y = format_output(train)
test_Y = format_output(test)


norm_train_X = norm(train)
norm_test_X = norm(test)
```

## Build & Trainig Model :

<div style="border: 5px solid yellow; padding: 10px; background-color: #F5F5F5; color: black; border-radius: 30px 45px 30px 45px;">

<p>&bull; Here is how we'll build the model using the functional syntax.</p>

<p>&bull; Notice that we can specify a list of outputs (i.e. `[y1_output, y2_output]`) when we instantiate the `Model()` class 

<p> &bull; at first ,Input Layer: The input shape is x_train.shape[1]  </p>

<p> &bull; Hidden Layers: </p> 

<p> &bull; Layer 1: Dense(128, activation="relu") with  Dropout(0.1).</p> 

<p> &bull; Layer 2: Dense(128, activation="relu") with  Dropout(0.1).</p>

<p> &bull; Y1 output will be fed directly from the second dense </p>

<p> &bull; Layer 3: Dense(128, activation="relu") with  Dropout(0.1) </p>

<p> &bull; Y2 output will be fed directly from the layer3 </p>

<p> &bull; We use Dropout layer to Reduce & prevent from overfiting .</p>

<p> &bull; Define the model with the input layer and a list of output layers </p>

<p> &bull; i used Gradient Descent optimizer ,with RootMeanSquare Error </p>

<p> &bull; i trained this model on 1000 epochs  </p>

</div>

```python
# Define a Callback class that stops training once accuracy reaches 98%

class myCallback(Callback):
  def on_epoch_end(self, epoch, logs={}):
    if(logs.get('loss')<5):
      print("\n Reached  5 % accuracy so cancelling training!")
      self.model.stop_training = True
```

```python
# build model :

# input_layer
input_layer=Input(shape=(8,))

# 1st layer
first_layer = Dense(128, activation="relu")(input_layer)
first_layer = Dropout(0.1)(first_layer)

# 2nd layer
second_layer = Dense(128, activation="relu")(first_layer)
second_layer = Dropout(0.1)(second_layer)

# output_1_layer : 
y1_output=Dense(1,name="y1_output")(second_layer)

# 3rd layer :
third_layer=Dense(128,activation="relu")(second_layer)
third_layer=Dropout(0.1)(second_layer)

# output_2_layer :

y2_output=Dense(1,name="y2_output")(third_layer)

# ApI_function :

model=Model(inputs=input_layer,outputs=[y1_output,y2_output])

print(model.summary())

# Compile the model :

optimizer = tf.keras.optimizers.SGD(learning_rate=0.001)
model.compile(optimizer=optimizer,
              loss={'y1_output': 'mse', 'y2_output': 'mse'},
              metrics={'y1_output': tf.keras.metrics.RootMeanSquaredError(),
                       'y2_output': tf.keras.metrics.RootMeanSquaredError()})

# train the model :

callbacks =myCallback()

print("\n Model_Training:")

history = model.fit(norm_train_X, train_Y,
                    epochs=1000, batch_size=100, validation_data=(norm_test_X, test_Y),callbacks=[callbacks])
```
